# Introduction

This is the erythroid lineage of gastrulation data from Pijuan-Sala et al. (2019) <https://doi.org/10.1038/s41586-019-0933-9>. The dataset is used in scvelo.dataset.gastrulation_erythroid and also in Barile et al. (2020) <https://doi.org/10.1186/s13059-021-02414-y>. The h5ad file `erythroid_lineage.h5ad` was downloaded from <https://ndownloader.figshare.com/files/27686871>
