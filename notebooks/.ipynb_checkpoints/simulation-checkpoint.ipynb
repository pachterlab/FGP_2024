# A simple case

We rescale the time by average half lives of mRNA

```python
#%% generate data
random_seed = 2022
loga_max = 4
logb_max = 3
n=1000
p=100
topo = np.array([[0,1]])
tau = (0,1,2)

np.random.seed(random_seed)
L=len(topo)
n_states=len(set(topo.flatten()))
t=np.linspace(tau[0],tau[-1],n)
true_t = []

theta=np.zeros((p,n_states+3))
for j in range(n_states+1):
    theta[:,j]=np.exp(np.random.uniform(0,loga_max,size=p))-1
theta[:,-2:]=np.exp(np.random.uniform(0,logb_max,size=(p,2)))

Y = np.zeros((n*L,p,2))
for l in range(L):
    theta_l = np.concatenate((theta[:,topo[l]], theta[:,-3:]), axis=1)
    Y[l*n:(l+1)*n] = get_Y(theta_l,t,tau) # m*p*2
    true_t = np.append(true_t,t)

X = np.random.poisson(Y)
```

```python
t = np.linspace(tau[0],tau[-1],n)
p = 10
fig, ax = plt.subplots(1,p,figsize=(6*p,4))
for i in range(p):
    #ax[i].plot(Y[:,i,1]*gamma[i],Y[:,i,1],'--', color='gray');
    ax[i].scatter(X[:,i,0],X[:,i,1],c='gray');
    ax[i].scatter(Y[:,i,0],Y[:,i,1],c=true_t);
# time increases from blue to red
```

```python
traj = Trajectory(topo, tau, model="two_species_ss", verbose=1)
traj = traj.fit(X,m=100,n_init=1,epoch=10,parallel=True,n_threads=10)
```

```python
fig, ax = plt.subplots(1,1)

for i in range(6):
    ax.scatter(np.arange(5,len(traj.elbos[i])), traj.elbos[i][5:],s=1)
```

```python
plot_phase(traj,idx=np.arange(10))
```

```python
plot_theta(theta, traj.theta)
```

```python
plot_t(traj.Q,l=0,t=true_t)
```

# A comple trajectory

```python
#%% generate data
random_seed = 2022
loga_max = 4
logb_max = 3
n=1000
p=100
topo = np.array([[0,1,2],[0,1,3]])
tau = (0,1,2,3)

np.random.seed(random_seed)
L=len(topo)
n_states=len(set(topo.flatten()))
t=np.linspace(tau[0],tau[-1],n)
true_t = []

theta=np.zeros((p,n_states+3))
for j in range(n_states):
    theta[:,j]=np.exp(np.random.uniform(0,loga_max,size=p))-1
theta[:,-3]=theta[:,0]
theta[:,-2:]=np.exp(np.random.uniform(0,logb_max,size=(p,2)))

Y = np.zeros((n*L,p,2))
for l in range(L):
    theta_l = np.concatenate((theta[:,topo[l]], theta[:,-3:]), axis=1)
    Y[l*n:(l+1)*n] = get_Y(theta_l,t,tau) # m*p*2
    true_t = np.append(true_t,t)

X = np.random.poisson(Y)
```

```python
t = np.linspace(tau[0],tau[-1],n)
p = 10
fig, ax = plt.subplots(1,p,figsize=(6*p,4))
for i in range(p):
    #ax[i].plot(Y[:,i,1]*gamma[i],Y[:,i,1],'--', color='gray');
    ax[i].scatter(X[:,i,0],X[:,i,1],c='gray');
    ax[i].scatter(Y[:,i,0],Y[:,i,1],c=true_t);
# time increases from blue to red
```

```python
traj = Trajectory(topo, tau, model="two_species_ss", verbose=1)
traj = traj.fit(X,m=100,n_init=2,epoch=100,parallel=True,n_threads=10)
```

```python
fig, ax = plt.subplots(1,1)

for i in range(len(traj.elbos)):
    ax.scatter(np.arange(len(traj.elbos[i])), traj.elbos[i],s=0.1)
```

```python
plt.plot(traj.elbos[0][80:101],'.')
plt.yscale('log')
```

```python
plot_phase(traj,idx=np.arange(10))
```

```python
plot_theta(theta, traj.theta)
```
