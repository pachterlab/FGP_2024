```python
subdata = adata
```

```python
sc.pp.highly_variable_genes(subdata,n_top_genes=2000,flavor="seurat_v3")
gene_mask = (subdata.var['highly_variable']) & (subdata.layers["unspliced"].toarray().mean(axis=0) > 0.5) & (subdata.layers["spliced"].toarray().mean(axis=0) > 1)
fitted_genes = subdata.var_names[gene_mask]
print(np.sort(fitted_genes))
```

marker genes: Piwil1 (spermatocytes), Tex21 (round spermatids), Tnp1 (elongating spermatids), Cldn11 (Sertoli cells) and Fabp3 (Leydig cells).



```python
marker_genes = ['Piwil1', 'Tex21','Tnp1','Cldn11', 'Fabp3']
plot_idx = np.arange(len(fitted_genes))[fitted_genes.isin(marker_genes)]
fitted_genes[plot_idx]
```

```python
X = np.zeros((subdata.X.shape[0],np.sum(gene_mask),2))
X[:,:,0] = subdata.layers["unspliced"].toarray()[:,gene_mask]
X[:,:,1] = subdata.layers["spliced"].toarray()[:,gene_mask]
X.shape
```

```python
topo = np.array([[0,1]])
tau = (0,50,100)
traj = Trajectory(topo, tau)
Q, elbo = traj.fit(X,parallel=True,n_threads=4)
```

```python
plt.scatter(X_umap[:,0],X_umap[:,1],c=Q[:,0]@np.linspace(0,2,100));
```

```python
plot_phase(X,Q,traj,idx=plot_idx,gene_name=fitted_genes)
```

```python
plot_y(X,Q,traj,idx=plot_idx,gene_name=fitted_genes)
```
