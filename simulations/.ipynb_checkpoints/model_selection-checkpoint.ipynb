```python
true_topo = np.array([[0,1,2]])
true_tau=(0,0.75,2)
n = 10000
p = 200
loga_mu=2
loga_sd=1
logb_mu=1
logb_sd=0.5
rd_var=0.25     

n_states=len(set(true_topo.flatten()))
t_samples=np.random.uniform(true_tau[0],true_tau[-1],size=100*n)

np.random.seed(0)                
theta=np.zeros((p,n_states+2))
for j in range(n_states):
    theta[:,j]=np.random.lognormal(loga_mu, loga_sd,size=p)
theta[:,-2:]=np.random.lognormal(logb_mu,logb_sd,size=(p,2))
theta[:,:n_states]/=theta[:,-2,None]

true_t = []
Y = np.zeros((n*L,p,2)) 
for l in range(L):
    t = np.random.choice(t_samples,n)
    theta_l = np.concatenate((theta[:,true_topo[l]], theta[:,-2:]), axis=1)
    Y[l*n:(l+1)*n] = get_Y(theta_l,t,true_tau) # m*p*2
    true_t = np.append(true_t,t)

if rd_var > 0:
    read_depth = np.random.gamma(1/rd_var,rd_var,n*L)
    rd = read_depth/read_depth.mean()
else:
    rd = np.ones(n*L)

X = np.random.poisson(rd[:,None,None]*Y)      
```

```python
topo_1 = np.array([[0,1,2]])
tau_1=(0,1.0,2)

topo_2 = np.array([[0,1,2,3]])
tau_2 = (0,0.1,1.0,2)

params={'r':rd,"batch_size":1000}
traj0_mini = Trajectory(true_topo, true_tau, model="two_species_ss", verbose=1)
traj0_mini = traj0_mini.fit(X,params=params,fit_tau=False,n_init=10,epoch=100,parallel=True,n_threads=4)

params={'r':rd}
traj0 = Trajectory(true_topo, true_tau, model="two_species_ss", verbose=1)
traj0 = traj0.fit(X,params=params,fit_tau=False,n_init=10,epoch=100,parallel=True,n_threads=4)
```

## Generate data and fit models

```python
true_topo = np.array([[0,1,2]])
true_tau=(0,0.75,2)

topo_1 = np.array([[0,1,2]])
tau_1=(0,1.0,2)

topo_2 = np.array([[0,1,2,3]])
tau_2 = (0,0.1,1.5,2)
           
def fit_model_wrapper(X, Y, true_t, theta, params):
    traj0 = Trajectory(topo_1, tau_1, model="two_species_ss", verbose=0)
    traj0 = traj0.fit(X,params=params,fit_tau=True,n_init=10,epoch=100,parallel=False,n_threads=1)
    traj0.true_t = true_t
    traj0.true_topo = true_topo
    traj0.true_tau = true_tau
    traj0.true_theta = theta
    traj0.Y = Y

    traj1 = Trajectory(topo_1, tau_1, model="two_species", verbose=0)
    traj1 = traj1.fit(X,params=params,fit_tau=True,n_init=10,epoch=100,parallel=False,n_threads=1)
    traj1.true_topo = true_topo
    traj1.true_tau = true_tau
    traj1.true_t = true_t
    traj1.true_theta = theta
    traj1.Y = Y

    traj2 = Trajectory(topo_2, tau_2, model="two_species_ss", verbose=0)
    traj2 = traj2.fit(X,params=params,fit_tau=True,n_init=10,epoch=100,parallel=False,n_threads=1)
    traj2.true_t = true_t
    traj2.true_topo = true_topo
    traj2.true_tau = true_tau
    traj2.true_theta = theta
    traj2.Y = Y

    return (traj0,traj1,traj2)

from multiprocessing import Pool

def vary_model_topo1(n=2000, p=200, t_samples=None, K=20, n_init=10, epoch=100, m=100, cores=15, loga_mu=2, loga_sd=1, logb_mu=1, logb_sd=0.5, rd_var=0.25):        
    L=len(true_topo)
    n_states=len(set(true_topo.flatten()))
    if t_samples is None:
        t_samples=np.random.uniform(true_tau[0],true_tau[-1],size=100*n)

    Input_args = []
    for k in range(K):
        np.random.seed(k)                
        theta=np.zeros((p,n_states+2))
        for j in range(n_states):
            theta[:,j]=np.random.lognormal(loga_mu, loga_sd,size=p)
        theta[:,-2:]=np.random.lognormal(logb_mu,logb_sd,size=(p,2))
        theta[:,:n_states]/=theta[:,-2,None]

        true_t = []
        Y = np.zeros((n*L,p,2)) 
        for l in range(L):
            t = np.random.choice(t_samples,n)
            theta_l = np.concatenate((theta[:,true_topo[l]], theta[:,-2:]), axis=1)
            Y[l*n:(l+1)*n] = get_Y(theta_l,t,true_tau) # m*p*2
            true_t = np.append(true_t,t)

        if rd_var > 0:
            read_depth = np.random.gamma(1/rd_var,rd_var,n*L)
            rd = read_depth/read_depth.mean()
        else:
            rd = np.ones(n*L)

        X = np.random.poisson(rd[:,None,None]*Y)      

        params={'r':rd}
        Input_args.append((X, Y, true_t, theta, params))
    
    with Pool(cores) as pool:      
        results = pool.starmap(fit_model_wrapper, Input_args, chunksize=1)
    
    return results
```

```python
results = vary_model_topo1(n=10000, p=200, t_samples=None, K=100,\
                           n_init=10, epoch=100, m=100, cores=50,\
                           loga_mu=2, loga_sd=1, logb_mu=1, logb_sd=0.5, rd_var=0)
with open(results_dir+'topo_1_10000.pickle', 'wb') as f:
    pickle.dump(results, f, pickle.HIGHEST_PROTOCOL)
```

```python
true_topo = np.array([[0,0,1,2],[0,0,1,3]])
true_tau = (0,0.75,2,3.0)

tau = (0,1,2,3.0)
topo_1 = np.array([[0,1,2,3],[0,1,2,4]])
topo_2 = np.array([[0,1,2,3],[0,1,4,5]])

def fit_model_wrapper_ws(X, Y, true_t, theta, params):
    traj1 = Trajectory(topo_1, tau, model="two_species_ss", verbose=0)
    traj1 = traj1.fit(X,params=params,fit_tau=True,n_init=10,epoch=100,parallel=False,n_threads=1)
    Q0 = traj1.Q.copy()

    traj0 = Trajectory(true_topo, tau, model="two_species_ss", verbose=0)
    traj0 = traj0.fit(X,warm_start=True,Q=Q0,params=params,fit_tau=True,n_init=10,epoch=100,parallel=False,n_threads=1)
    traj0.true_t = true_t
    traj0.true_topo = true_topo
    traj0.true_tau = true_tau
    traj0.true_theta = theta
    traj0.Y = Y
   
    traj1 = Trajectory(topo_1, tau, model="two_species_ss", verbose=0)
    traj1 = traj1.fit(X,warm_start=True,Q=Q0,params=params,fit_tau=True,n_init=10,epoch=100,parallel=False,n_threads=1)

    traj2 = Trajectory(topo_1, tau, model="two_species", verbose=0)
    traj2 = traj2.fit(X,warm_start=True,Q=Q0,params=params,fit_tau=True,n_init=10,epoch=100,parallel=False,n_threads=1)
    
    traj3 = Trajectory(topo_2, tau, model="two_species_ss", verbose=0)
    traj3 = traj3.fit(X,warm_start=True,Q=Q0,params=params,fit_tau=True,n_init=10,epoch=100,parallel=False,n_threads=1)
    
    return (traj0,traj1,traj2,traj3)

def fit_model_wrapper(X, Y, true_t, theta, params):
    traj0 = Trajectory(true_topo, tau, model="two_species_ss", verbose=0)
    traj0 = traj0.fit(X,params=params,fit_tau=True,n_init=10,epoch=100,parallel=False,n_threads=1)
    traj0.true_t = true_t
    traj0.true_topo = true_topo
    traj0.true_tau = true_tau
    traj0.true_theta = theta
    traj0.Y = Y
   
    traj1 = Trajectory(topo_1, tau, model="two_species_ss", verbose=0)
    traj1 = traj1.fit(X,params=params,fit_tau=True,n_init=10,epoch=100,parallel=False,n_threads=1)

    traj2 = Trajectory(topo_1, tau, model="two_species", verbose=0)
    traj2 = traj2.fit(X,params=params,fit_tau=True,n_init=10,epoch=100,parallel=False,n_threads=1)
    
    traj3 = Trajectory(topo_2, tau, model="two_species_ss", verbose=0)
    traj3 = traj3.fit(X,params=params,fit_tau=True,n_init=10,epoch=100,parallel=False,n_threads=1)
    
    return (traj0,traj1,traj2,traj3)

from multiprocessing import Pool

def vary_model_topo2(ws, n=2000, p=200, t_samples=None, K=20, n_init=10, epoch=100, m=100, cores=15, loga_mu=2, loga_sd=1, logb_mu=1, logb_sd=0.5, rd_var=0.25):        
    L=len(true_topo)
    n_states=len(set(true_topo.flatten()))
    if t_samples is None:
        t_samples=np.random.uniform(true_tau[0],true_tau[-1],size=100*n)

    Input_args = []
    for k in range(K):
        np.random.seed(k)                
        theta=np.zeros((p,n_states+2))
        for j in range(n_states):
            theta[:,j]=np.random.lognormal(loga_mu, loga_sd,size=p)
        theta[:,-2:]=np.random.lognormal(logb_mu,logb_sd,size=(p,2))
        theta[:,:n_states]/=theta[:,-2,None]

        true_t = []
        Y = np.zeros((n*L,p,2)) 
        for l in range(L):
            t = np.random.choice(t_samples,n)
            theta_l = np.concatenate((theta[:,true_topo[l]], theta[:,-2:]), axis=1)
            Y[l*n:(l+1)*n] = get_Y(theta_l,t,true_tau) # m*p*2
            true_t = np.append(true_t,t)

        if rd_var > 0:
            read_depth = np.random.gamma(1/rd_var,rd_var,n*L)
            rd = read_depth/read_depth.mean()
        else:
            rd = np.ones(n*L)

        X = np.random.poisson(rd[:,None,None]*Y)      

        params={'r':rd}
        Input_args.append((X, Y, true_t, theta, params))
    
    with Pool(cores) as pool:     
        if ws:
            results = pool.starmap(fit_model_wrapper_ws, Input_args, chunksize=1)
        else:
            results = pool.starmap(fit_model_wrapper, Input_args, chunksize=1)
            
    
    return results
```

```python
results = vary_model_topo2(ws=True,n=2000, p=200, t_samples=None, K=100,\
                           n_init=10, epoch=100, m=100, cores=20,\
                           loga_mu=2, loga_sd=1, logb_mu=1, logb_sd=0.5, rd_var=0)
with open(results_dir+'topo_2_ws_2000.pickle', 'wb') as f:
    pickle.dump(results, f, pickle.HIGHEST_PROTOCOL)
```
